```python
import pandas as pd
pd.options.display.max_rows = 10

import sqlite3
conn = sqlite3.connect('task1_7.db')  
c = conn.cursor()
```

```python
request = """
create table table1
(
  x int not null,
  y int not null,
  z int not null
);
"""
c.execute(request)
```

```python
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", conn)
```

```python
request = """
INSERT INTO table1
    WITH RECURSIVE
      cnt(x, y, z) AS (
      VALUES(abs(random()%1001), abs(random()%1001), abs(random()%1001)) UNION ALL 
      SELECT abs(random()%1001),abs(random()%1001), abs(random()%1001) FROM cnt LIMIT 1000)
    select * from cnt;
"""
c.execute(request)
```

```python
pd.read_sql_query("SELECT * FROM table1", conn)
```
