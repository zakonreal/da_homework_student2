```python
tim = [item['created_at'] for item in vacs]
tim
```

```python
v_t = dict(zip(vact,tim))
v_t
```

```python
from collections import Counter
```

```python
cnt_vact = Counter(list(vact))
```

```python
cnt_tim = Counter(list(tim))
```

```python
cnt_s = Counter(v_t)
```

```python
# число всех посчитанных элементов
sum(cnt_vact.values())
```

```python
# Подсчет количества повторений элементов в последовательности
dict(cnt_vact)
```

```python
#список наиболее распространенных элементов
cnt_vact.most_common(10)
```

```python
# сортировка
sorted(cnt_vact)
```

```python
# сортировка
from collections import OrderedDict
y = OrderedDict(cnt_vact.most_common())
y
```
